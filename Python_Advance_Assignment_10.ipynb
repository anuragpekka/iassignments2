<b>2. What is the difference between properties and descriptors?</b><br>Answer:<br>
- The <b>property()</b> function is used to define properties in the Python class.The property() method in Python provides an interface to instance attributes. It encapsulates instance attributes and provides a property.
It takes four optional parameters:<br>
&emsp;fget (optional) - Function for getting the attribute value. Defaults to None.<br>
&emsp;fset (optional) - Function for setting the attribute value. Defaults to None.<br>
&emsp;fdel (optional) - Function for deleting the attribute value. Defaults to None.<br>
&emsp;doc (optional) - A string that contains the documentation (docstring) for the attribute. Defaults to None.<br><br>
Return value from property()<br>
property() returns the property attribute from the given getter, setter, and deleter. If no arguments are given, property() returns a base property attribute that doesn't contain any getter, setter or deleter. If doc isn't provided, property() takes the docstring of the getter function.<br>
Example 1: Create attribute with getter, setter, and deleter<br>
class Person:<br>
&emsp;def &#95;&#95;init&#95;&#95;(self, name):
<br>&emsp;&emsp;self.&#95;&#95;name = name
<br>&emsp;def get_name(self):
<br>&emsp;&emsp;print('Getting name')
<br>&emsp;&emsp;return self.&#95;name
<br>
<br>&emsp;def set_name(self, value):
<br>&emsp;&emsp;print('Setting name to ' + value)
<br>&emsp;&emsp;self.&#95;name = value
<br>
<br>&emsp;def del_name(self):
<br>&emsp;&emsp;print('Deleting name')
<br>&emsp;&emsp;del self.&#95;name
<br>
<br>&emsp;# Set property to use get_name, set_name
<br>&emsp;# and del_name methods
